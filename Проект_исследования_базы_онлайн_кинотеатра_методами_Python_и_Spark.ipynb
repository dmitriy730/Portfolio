1. Загружаем в колаб файлы по оценкам (ratings) и фильмам (movies) и создаем на их основе pandas-датафреймы

```python
import pandas as pd
```

```python
# Загружаем файл по рейтингам
df = pd.read_csv('rating.csv')
df.head()
```

```python
len(df)
```

```python
# Загружаем файл по фильмам
df_f = pd.read_csv('movies.csv')
df_f.head()
```

```python
len(df_f)
```

2. Средствами Pandas, используя dataframe ratings, находим id пользователя, поставившего больше всего оценок



```python
df_1 = df['ID user'].value_counts().reset_index().head(1)
df_1
```

Пользователь с ID = 405 поставил больше всего оценок - 737

3. Оставляем в датафрейме ratings только те фильмы, который оценил данный пользователь

```python
df_2 = df[(df['ID user'] == 405)]
df_2
```

4. Добавляем к датафрейму из задания 3 столбцы:
- По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет.
- Столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей


```python
# Объединяем таблицы рейтингов и фильмов левым соединением по ключу ID film
df_3 = df_2.merge(df_f, how='left', on='ID film')
df_3.head()
```

```python
# Общее количество оценок и суммарная оценка от всех пользователей на фильм
df_4 = df.groupby('ID film').agg(['count', 'sum'])['rating'].reset_index()
df_4.head()
```

```python
# Cоединяем снова полученные данные
df_5 = df_3.merge(df_4, how='left', on='ID film')
df_5.head()
```

Разбиваем данные на множество для обучения и для проверки

Берем модель линейной регрессии и обучаем ее



```python
# Выводим названия столбцов датафрейма
df_5.columns
```

```python
df_lda = df_5[['Неизвестен', 'Боевик', 'Приключения', 'Анимация', 'Детский', 'Комедия', 'Криминал', 'Документальный', 'Драма',
       'Фентези', 'Нуар', 'Ужасы', 'Мюзикл', 'Мистика', 'Романтика',
       'Фантастика', 'Триллер', 'Война', 'Вестерн', 'count', 'sum']]
```

```python
df_class = df_5['rating']
```

```python
# Импортируем библиотеку
import sklearn
```

```python
# разделяем выборку на тренировочную и тестовую
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_lda, df_class, random_state=40)
```

```python
# импортируем LDA-функционал
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
# обучаем данные
lda.fit(X_train, y_train)
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

```python
# смотрим разницу факта и прогноза
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
```

```python
result
```

```python
# Оценим качество модели метрикой точности
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, lda.predict(X_test))
```

С учетом имеющихся данных, модель, предсказывающая оценку фильма пользователем имеет точность 61% на тестовой выборке.
